```python
import numpy as np
import matplotlib.pyplot as plt
#PCB1 = not fixed
#PCB2 = fixed
```

```python
def find_index(value, arr):
    i=0
    found = False
    while found==False:
        if (arr[i]<=value<arr[i+1]) or (arr[i]>=value>arr[i+1]):
            found = True
            idx = i
        i+=1
    return idx
```

```python
#CLK signal PCB1
data1 = np.genfromtxt('./data/pcb1_clk.csv', dtype=float, skip_header=2, delimiter=',')

fig = plt.figure(figsize=(8,5))
plt.plot(data1[:,0], data1[:,1])
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel("t [s]", fontsize=14)
plt.ylabel("Voltage [V]", fontsize=14)
plt.grid()
#plt.savefig('./clk_signal_PCB1.eps', bbox_inches='tight')
plt.show()

data2 = np.genfromtxt('./data/pcb2_clk.csv', dtype=float, skip_header=2, delimiter=',')
fig = plt.figure(figsize=(8,5))
plt.plot(data2[:,0], data2[:,1])
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel("t [s]", fontsize=14)
plt.ylabel("Voltage [V]", fontsize=14)
plt.grid()
#plt.savefig('clk_signal_PCB2.eps',bbox_inches='tight')
plt.show()
```

```python
# or input and output @ min carry
data1 = np.genfromtxt('./data/pcb2_or_input.csv', dtype=float, skip_header=2, delimiter=',')
data2 = np.genfromtxt('./data/pcb2_or_output.csv', dtype=float, skip_header=2, delimiter=',')
fig = plt.figure(figsize=(8,5))
plt.plot(data1[:,0], data1[:,1], label='Input')
plt.plot(data2[:,0], data2[:,1], label='Output')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel("t [s]", fontsize=14)
plt.ylabel("Voltage [V]", fontsize=14)
plt.grid()
plt.legend()

data1 = np.genfromtxt('./data/pcb1_or_input.csv', dtype=float, skip_header=2, delimiter=',')
data2 = np.genfromtxt('./data/pcb1_or_output.csv', dtype=float, skip_header=2, delimiter=',')
fig = plt.figure(figsize=(8,5))
plt.plot(data1[:,0], data1[:,1], label='Input')
plt.plot(data2[:,0], data2[:,1], label='Output')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel("t [s]", fontsize=14)
plt.ylabel("Voltage [V]", fontsize=14)
plt.grid()
plt.legend()
```

```python
# or input and output @ min carry --> ascending
data1 = np.genfromtxt('./data/pcb1_or_input_ascending.csv', dtype=float, skip_header=2, delimiter=',')
data2 = np.genfromtxt('./data/pcb1_or_output_ascending.csv', dtype=float, skip_header=2, delimiter=',', usecols=[1])
data_pcb1 = np.column_stack( (data1, data2) )

data1 = np.genfromtxt('./data/pcb2_or_input_ascending.csv', dtype=float, skip_header=2, delimiter=',')
data2 = np.genfromtxt('./data/pcb2_or_output_ascending.csv', dtype=float, skip_header=2, delimiter=',', usecols=[1])
data_pcb2 = np.column_stack( (data1, data2) )

idx1 = find_index(1,data_pcb1[:,2])
idx2 = find_index(1,data_pcb2[:,2])
data_pcb2[:,0] += data_pcb1[idx1,0] - data_pcb2[idx2,0]
x_idx1 = find_index(0.004, data_pcb1[:,0])
x_idx2 = find_index(0.0048, data_pcb1[:,0])

fig = plt.figure(figsize=(8,5))
plt.plot(data_pcb1[:,0], data_pcb1[:,1], label='OR-Input, PCB1', alpha=0.7)
plt.plot(data_pcb1[:,0], data_pcb1[:,2], label='OR-Output, PCB1', alpha=0.7)
plt.plot(data_pcb2[:,0], data_pcb2[:,1], label='OR-Input, PCB2', alpha=0.7)
plt.plot(data_pcb2[:,0], data_pcb2[:,2], label='OR-Output, PCB2', alpha=0.7)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('t [s]', fontsize=14)
plt.ylabel('Voltage [V]', fontsize=14)
plt.xlim([data_pcb1[x_idx1,0], data_pcb1[x_idx2,0] ])
plt.grid()
plt.legend()
#plt.savefig('ascending_OR.pdf', bbox_inches='tight')
plt.show()
```

```python
# or input and output @ min carry --> ascending
data1 = np.genfromtxt('./data/pcb1_or_input_descending.csv', dtype=float, skip_header=2, delimiter=',')
data2 = np.genfromtxt('./data/pcb1_or_output_descending.csv', dtype=float, skip_header=2, delimiter=',', usecols=[1])
data_pcb1 = np.column_stack( (data1, data2) )

data1 = np.genfromtxt('./data/pcb2_or_input_descending.csv', dtype=float, skip_header=2, delimiter=',')
data2 = np.genfromtxt('./data/pcb2_or_output_descending.csv', dtype=float, skip_header=2, delimiter=',', usecols=[1])
data_pcb2 = np.column_stack( (data1, data2) )

idx1 = find_index(4, data_pcb1[:,2])
idx2 = find_index(4, data_pcb2[:,2])
data_pcb2[:,0] += data_pcb1[idx1,0] - data_pcb2[idx2,0]
x_idx1 = find_index(-0.0013, data_pcb1[:,0])
x_idx2 = find_index(0.0013, data_pcb1[:,0])

fig = plt.figure(figsize=(8,5))
plt.plot(data_pcb1[:,0], data_pcb1[:,1], label='OR-Input, PCB1', alpha=0.7)
plt.plot(data_pcb1[:,0], data_pcb1[:,2], label='OR-Output, PCB1', alpha=0.7)
plt.plot(data_pcb2[:,0], data_pcb2[:,1], label='OR-Input, PCB2', alpha=0.7)
plt.plot(data_pcb2[:,0], data_pcb2[:,2], label='OR-Output, PCB2', alpha=0.7)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('t [s]', fontsize=14)
plt.ylabel('Voltage [V]', fontsize=14)
plt.xlim([data_pcb1[x_idx1,0], data_pcb1[x_idx2,0] ])
plt.grid()
plt.legend()
#plt.savefig('descending_OR.pdf', bbox_inches='tight')
plt.show()
```
